```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load your dataset
df = pd.read_csv('dataset\\the-reddit-dataset-dataset-comments.csv')
```

```python
# Check for sentiment distribution
sns.histplot(df['sentiment'], kde=True, bins=20, color='skyblue')
plt.title('Sentiment Distribution in the Subreddit')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot to see the relationship between sentiment and Reddit score
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='sentiment', y='score')
plt.title('Sentiment Score vs Reddit Score in the Subreddit')
plt.xlabel('Sentiment Score')
plt.ylabel('Reddit Score')
plt.show()
```

```python
# Display comments with the highest positive sentiment
print("Top 5 Positive Comments:")
print(df.sort_values(by='sentiment', ascending=False)[['body', 'sentiment']].head())

# Display comments with the highest negative sentiment
print("Top 5 Negative Comments:")
print(df.sort_values(by='sentiment', ascending=True)[['body', 'sentiment']].head())
```

```python
# Boxplot to show the distribution of Reddit scores for different sentiment categories
df['sentiment_category'] = df['sentiment'].apply(lambda x: 'positive' if x > 0 else 'negative' if x < 0 else 'neutral')

sns.boxplot(x='sentiment_category', y='score', data=df)
plt.title('Reddit Score Distribution by Sentiment Category')
plt.xlabel('Sentiment Category')
plt.ylabel('Reddit Score')
plt.show()
```
